```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sb.set()
```

```python
df = pd.read_csv('disease_prediction.csv')
```

```python
df.head()
```

```python
df.describe(include = 'all')
```

```python
new_df = df
```

```python
new_df = df.drop(['patient_id'], axis = 1)
```

```python
new_df.head(10)
```

```python
new_df.columns
```

```python
df_cols = ['gender', 'smoking', 'alcohol_consumption', 'family_history', 'disease']
```

```python
for col in df_cols:
    if 'Male' in new_df[col].unique():
        new_df[col] = new_df[col].map({'Male' : 1, 'Female' : 0})
    else:
        new_df[col] = new_df[col].map({'Yes' : 1, 'No' : 0})
    
```

```python
new_df['physical_activity'] = new_df['physical_activity'].map({'Low' : 0, 'Medium' : 1, 'High' : 2})
```

```python
new_df.head()
```

```python
##Split the Data
```

```python
inputs = new_df.drop(['disease'], axis = 1)
target = new_df['disease']
```

```python
inputs
```

```python
target
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs, target, test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
numeric_cols = ['age', 'glucose_mg_dl', 'cholesterol_mg_dl', 'systolic_bp', 'diastolic_bp', 'bmi', 'heart_rate']
```

```python
scaler = StandardScaler()
```

```python
x_train[numeric_cols] = scaler.fit_transform(x_train[numeric_cols])
x_test[numeric_cols] = scaler.transform(x_test[numeric_cols])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state = 42)
model.fit(x_train, y_train)
print("Model Training Successful")
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
df_compare = pd.DataFrame({
    'Actual Disease' : y_test.values,
    'Prediction' : y_pred,
    
})
df_compare.head(15)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(
    cm,
    index = ['Actual_Healthy (0)', 'Actual_Disease (1)'],
    columns = ['Predicted_Healthy (0)', 'Actual_Disease (1)']
)
```

```python
cm_df
```

```python
print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred))
```
